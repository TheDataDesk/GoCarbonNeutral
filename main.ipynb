```python
#Load Data 

import pandas as pd 

file_path ='/Users/siri/BVGCN.github.io/data.json'

df = pd.read_json(file_path)

# Display the DataFrame
print(df)
```

```python
#For alexander we are only trying to determine from Berlin Zoo to  East Side Gallery
# Drop the row with index 1
df = df.drop(1, axis=0)

# Display the DataFrame after dropping the row
print(df)
```

```python
# Flatten 'fromStops' and 'toStops'
df_from_stops = pd.json_normalize(df['fromStops'].explode().tolist())
df_to_stops = pd.json_normalize(df['toStops'].explode().tolist())

# Flatten 'journeyData'
df_journey_data = pd.json_normalize(df['journeyData'])

# Drop the original nested columns
df_clean = df.drop(['fromStops', 'toStops', 'journeyData'], axis=1)

# Combine flattened data with original DataFrame
df_final = pd.concat([df_clean, df_from_stops, df_to_stops, df_journey_data], axis=1)

# Display the final DataFrame
print(df_final)
```
